```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data.csv', sep=',')

melb_df = melb_data.copy()

melb_df['Date'] = pd.to_datetime(melb_df['Date'], format='%d/%m/%Y')

melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
melb_df = melb_df.drop('YearBuilt', axis=1)

melb_df['WeekdaySale'] = melb_df['Date'].dt.day_of_week
weekend_count = melb_df[(melb_df['WeekdaySale'] == 5) | (melb_df['WeekdaySale'] == 6)].shape[0]

def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        return 1
    return 0

melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)


# отсавляес только 49 самых популярных компаний все остальные запмсываем как other 
comp = melb_df['SellerG'].value_counts().nlargest(49).index
melb_df['SellerG'] = melb_df['SellerG'].apply(lambda x: x if x in comp else 'other')
# во сколько минимальная цена Nelson отличается от минимальной цены other 
nelson_mean = melb_df[melb_df['SellerG'] == 'Nelson']['Price'].min()
other_mean = melb_df[melb_df['SellerG'] == 'other']['Price'].min()
print(nelson_mean / other_mean)
print(melb_df.memory_usage().sum() / 1024 / 1024)
prig = melb_df['Suburb'].value_counts().nlargest(119).index
melb_df['Suburb'] = melb_df['Suburb'].apply(lambda x: x if x in prig else 'other').astype('category')
print(melb_df.memory_usage().sum() / 1024 / 1024)
```

```python
# создаём пустой список
unique_list = []
# пробегаемся по именам столбцов в таблице
for col in melb_df.columns:
    # создаём кортеж (имя столбца, число уникальных значений)
    item = (col, melb_df[col].nunique(),melb_df[col].dtypes) 
    # добавляем кортеж в список
    unique_list.append(item) 
# создаём вспомогательную таблицу и сортируем её
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique',  ignore_index=True)
# выводим её на экран
display(unique_counts)
```

```python
import pandas as pd

bike_data = pd.read_csv('data/citibike-tripdata.csv', sep=',')
bike_df = bike_data.copy()
bike_df.info()
```

```python
bike_df['starttime'] = pd.to_datetime(bike_df['starttime'])
bike_df['stoptime'] = pd.to_datetime(bike_df['stoptime'])

bike_df['stoptime'].dt.year.max() - bike_df['birth year'].max()

bike_df['start station name'].value_counts()
bike_df.drop(['start station id', 'end station id'], axis=1, inplace=True)

bike_df['age'] = 2018 - bike_df['birth year']
bike_df.drop(['birth year'], axis=1, inplace=True)

# print(bike_df[bike_df['age'] > 60].shape)

bike_df['trip duration'] = bike_df['stoptime'] - bike_df['starttime']

def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        return 1
    return 0

bike_df['weekend'] = bike_df['starttime'].dt.day_of_week.apply(get_weekend)
#print(bike_df['weekend'].value_counts())


def get_timesofday(hour):
    # hour = time.dt.hour
    # 0-6 nigth 6-12 morning 12-18 day 18-23 evening
    if hour in range(7):
        return 'night'
    elif hour in range(7, 13):
        return 'morning'
    elif hour in range(13, 19):
        return 'day'
    return 'evening'

bike_df['time of day'] = bike_df['starttime'].dt.hour.apply(get_timesofday)

day_count = int(bike_df[bike_df['time of day'] == 'day'].shape[0])
night_count = int(bike_df[bike_df['time of day'] == 'night'].shape[0])

print('во соклько раз поездок днем больше чем ночью: ', day_count / night_count)

display(bike_df.head())
print(len(bike_df.columns))
```

```python
import pandas as pd 

ratings1 = pd.read_csv('data/ratings1.csv')
ratings2 = pd.read_csv('data/ratings2.csv')
movies = pd.read_csv('data/movies.csv')
dates = pd.read_csv('data/dates.csv')

# display(dates) # y-m-d hh:mm:ss
dates['date'] = pd.to_datetime(dates['date'], format='%Y-%m-%d %H:%M:%S')
# display(dates['date'].dt.year.value_counts())

ratings = pd.concat([ratings1, ratings2], ignore_index=True)
ratings = ratings.drop_duplicates(ignore_index=True)
# display(ratings)
ratings_dates = pd.concat([ratings, dates], axis=1)
# display(ratings_dates.tail(7))


joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)

display(joined)

#библиотека для регулярных выражений
import re 

def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
    
    
```

```python
joined['year_release'] = joined['title'].apply(get_year_release)
display(joined.head(3))
# display(joined[joined['year_release'] == 1999].groupby('title')['rating'].mean().sort_values())
# display(joined[joined['year_release'] == 2010].groupby('genres')['rating'].mean().sort_values())
# display(joined.groupby('userId')['genres'].nunique().sort_values())
# display(joined.groupby('userId')['rating'].agg(['mean', 'count']).sort_values(['count', 'mean'], ascending=[True, False]))
# fill = joined[joined['year_release'] == 2018].groupby('genres')['rating'].agg(['mean', 'count']).sort_values('mean', ascending=False)
# display(fill[(fill['count'] > 0)])
joined['year_rating'] = joined['date'].dt.year 
# display(joined)
pivot_rating = joined.pivot_table(values='rating', index='genres', columns='year_rating', fill_value=0)
pivot_rating.loc['Action|Adventure|Animation|Children|Comedy|IMAX'].idxmax() # индекс максимальной оценки
display(pivot_rating[2018].sort_values().tail(100)['Animation|Children|Mystery'])
```

```python
import pandas as pd 

covid_data = pd.read_csv('data/covid_data.csv')
# display(covid_data.head(5))
covid_vac = pd.read_csv('data/country_vaccinations.csv')
vaccinations_data = covid_vac[
    ['country', 'date', 'total_vaccinations', 
     'people_vaccinated', 'people_vaccinated_per_hundred',
     'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred',
     'daily_vaccinations', 'vaccines']
]
# display(vaccinations_data.head(5))

# Группируем таблицу по дате и названию страны и рассчитываем суммарные показатели по всем регионам. Тем самым переходим от данных по регионам к данным по странам:
covid_data = covid_data.groupby(
    ['date', 'country'], 
    as_index=False
)[['confirmed', 'deaths', 'recovered']].sum()

# Преобразуем даты в формат datetime с помощью функции pd.to_datetime()
covid_data['date'] = pd.to_datetime(covid_data['date'])

# Преобразуем даты в формат datetime с помощью функции pd.to_datetime()
covid_data['active'] = covid_data['confirmed'] - covid_data['deaths'] - covid_data['recovered']

# Создадим признак ежедневного прироста числа заболевших, умерших и выздоровевших людей. Для этого отсортируем данные по названиям стран, 
# а затем по датам. После этого произведём группировку по странам и рассчитаем разницу между «вчера и сегодня» с помощью метода diff():
covid_data = covid_data.sort_values(by=['country', 'date'])
covid_data['daily_confirmed'] = covid_data.groupby('country')['confirmed'].diff()
covid_data['daily_deaths'] = covid_data.groupby('country')['deaths'].diff()
covid_data['daily_recovered'] = covid_data.groupby('country')['recovered'].diff()

# В таблице vaccinations_data достаточно будет преобразовать столбцы в формат datetime 
vaccinations_data['date'] = pd.to_datetime(vaccinations_data['date'])


display(covid_data.sort_values(by='date'))
```
